### 1.5.6 원소 접근

```python
X = np.array([[51, 55], [14, 19], [0, 4]])
print(X)
```

```python
X[0]
```

```python
X[0][1]
```

```python
# for 문으로도 각 원소에 접근 가능
for row in X:
    print(row)
```

```python
X = X.flatten() # X를 1차원 배열로 변환(평탄화)
print(X)
```

```python
X[np.array([0, 2, 4])] # index를 배열로 지정해 한 번에 여러 원소로 접근이 가능하다
```

```python
X > 15
```

```python
X[X > 15]
```

## 1.6 matplotlib

### 1.6.1 단순한 그래프 그리기

```python
import numpy as np
import matplotlib.pyplot as plt

# 데이터 준비
x = np.arange(0, 6, 0.1) # 0부터 6까지 0.1 간격으로 설정
y = np.sin(x)

# 그래프 그리기
plt.plot(x, y)
plt.show()
```

### 1.6.2 pyplot의 기능

```python
x = np.arange(0, 6, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
plt.plot(x, y1, label="sin")
plt.plot(x, y2, linestyle="--", label="cos") # cos 함수는 점선으로 그리기
plt.xlabel("x") # x축 이름
plt.ylabel("y") # y축 이름
plt.title("sin & cos")
plt.legend()
plt.show()
```
